```python
# 1. Carga del dataset
df_consolidado = pd.read_csv(r'criptoradar_dataset_final.csv')
# 2. Visualizar las primeras 5 filas
print("--- Primeras filas del dataset ---")
display(df_consolidado.head())

# 3. Inspección de tipos de datos y valores no nulos
print("\n--- Información general del DataFrame ---")
df_consolidado.info()
```

```python
print("--- Correlación entre el Supply de la Stablecoin y cada Criptomoneda ---")

# Agrupamos por cada moneda y calculamos la correlación interna de sus columnas
correlaciones_por_crypto = df_consolidado.groupby('crypto_id').apply(
    lambda x: x['supply_stablecoin'].corr(x['precio_crypto'])
)

display(correlaciones_por_crypto)
```

Al ser un numero entre -0.4 y -0.9.Se demuestra con datos reales de 1 año que cuando el precio de Bitcoin baja, el suministro de stablecoins sube (los inversores buscan refugio)y eso se ve reflejado en el aumento de precio de supply.
